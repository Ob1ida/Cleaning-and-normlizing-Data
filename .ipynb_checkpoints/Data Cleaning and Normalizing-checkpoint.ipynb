```python
df = pd.read_csv('Ames_outliers_removed.csv')
```

```python
#drop this column its just uniqe ids we can get access from indexes
df = df.drop('PID',axis=1)
```

```python
df
```

```python
df.isnull().sum()
```

```python
#this func calculate the precentage of nan values for each column
def percent_missing(df):
    percent_nan = 100*df.isnull().sum()/len(df)
    percent_nan = percent_nan[percent_nan>0].sort_values()
    
    
    return percent_nan
```

```python
a = percent_missing(df)
```

```python
a
```

```python
#lets visualize this
sns.barplot(x=a.index,y=a)
plt.xticks(rotation = 90);
plt.ylim(0,1)
```

```python
#this is the columns that have less than 1% od Nan Values
a[a<1]
```

```python
#for Electrical and Bsmt Half Bath columns there are 2 rows that have Nan values
# we can simply drop them from dataframe
df = df.dropna(axis=0,subset=['Electrical','Garage Cars'])
```
